```python
import nltk, random
from nltk.corpus import reuters
import pandas as pd

# Download the dataset
nltk.download('reuters')

from sklearn.datasets import fetch_20newsgroups
import random
```

```python


# Load the dataset
newsgroups = fetch_20newsgroups(subset='all', categories=None)

# Randomly sample 100 documents
sample_indices = random.sample(range(len(newsgroups.data)), 100)
sampled_data = [newsgroups.data[i] for i in sample_indices]
sampled_labels = [newsgroups.target[i] for i in sample_indices]

print(f"Sampled Data: {sampled_data[:1]}")
print(f"Sampled Label: {newsgroups.target_names[sampled_labels[0]]}")
```

```python
# Convert to a DataFrame
df_20newsgroups = pd.DataFrame({
    'text': newsgroups.data,
    'target': newsgroups.target
})

# Map the numeric target to the descriptive topic name
df_20newsgroups['target_name'] = df_20newsgroups['target'].apply(
    lambda x: newsgroups.target_names[x]
)
print('df_20newsgroups.shape:', df_20newsgroups.shape)
print('df_20newsgroups.info():', df_20newsgroups.info())
df_20newsgroups.head()
```

```python
set_of_target_names = set(list(df_20newsgroups['target_name']))
print(len(set_of_target_names))
set_of_target_names
```

```python


# Get a list of documents
documents = reuters.fileids()
random.shuffle(documents)

# Sample 100 documents
sample_docs = documents[:100]
sample_data = [reuters.raw(doc_id) for doc_id in sample_docs]
sample_labels = [reuters.categories(doc_id) for doc_id in sample_docs]

print(f"Sampled Document: {sample_data[0]}")
print(f"Sampled Labels: {sample_labels[0]}")
```

```python
# Create a list of dictionaries, one dict per file/document
docs = []
for fileid in reuters.fileids():
    docs.append({
        "fileid": fileid,
        "text": reuters.raw(fileid),
        "categories": reuters.categories(fileid),
        # Optionally, you can include tokenized forms if needed:
        # "words": reuters.words(fileid),
        # "sentences": reuters.sents(fileid)
    })

# Convert the list of dicts into a DataFrame
df_reuters = pd.DataFrame(docs)

# Print a sample
print('df_reuters.shape:', df_reuters.shape)
print('df_reuters.info():', df_reuters.info())
df_reuters.head()
```

```python
import itertools

unique_categories = set(itertools.chain.from_iterable(df_reuters['categories']))
print(len(unique_categories))
unique_categories
```
